```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas here

# Load the dataset and clean it within this script
df = pd.read_csv("student_depression_dataset.csv")
df_cleaned = df.drop_duplicates()

# ... (Rest of your plotting code remains the same) ...

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 1. Bar Plot: Average depression score by gender
plt.figure()
sns.barplot(data=df_cleaned, x='Gender', y='Depression')
plt.title("Average Depression Score by Gender")
plt.ylabel("Depression Level")
plt.xlabel("Gender")
plt.show()

# 2. Box Plot: CGPA distribution by level of depression
plt.figure()
sns.boxplot(data=df_cleaned, x='Depression', y='CGPA')
plt.title("CGPA Distribution across Depression Levels")
plt.xlabel("Depression Level")
plt.ylabel("CGPA")
plt.show()

# 3. Heatmap: Correlation matrix of numerical features
plt.figure()
corr = df_cleaned.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

# 4. Scatter Plot: CGPA vs. Work/Study Hours colored by Depression
plt.figure()
sns.scatterplot(data=df_cleaned, x='Work/Study Hours', y='CGPA', hue='Depression', palette='cool')
plt.title("CGPA vs. Work/Study Hours by Depression Level")
plt.xlabel("Work/Study Hours")
plt.ylabel("CGPA")
plt.show()

# 5. Pie Chart: Distribution of dietary habits
plt.figure()
diet_counts = df_cleaned['Dietary Habits'].value_counts()
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Dietary Habits")
plt.axis('equal')
plt.show()
```
